```python
# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find attributes with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

to_drop

# Drop high correlated attributes 
# data.drop(to_drop, axis=1, inplace=True)
```

### Groupby, filter, merge and concat

```python
# Get the maximum of first period's grades by grouping school, age and gender
grp_g1 = data.groupby(['school', 'age', 'sex'])['G1'].max().reset_index(name='G1_max')
```

```python
grp_g1
```

```python
# Get the maximum of second period's grades by grouping school, age and gender
grp_g2 = data.groupby(['school', 'age', 'sex'])['G2'].max().reset_index(name='G2_max')
```

```python
grp_g2
```

```python
# Filter the school -- find Mousinho da Silveira (MS) school
grp_g1_ms = grp_g1.loc[grp_g1['school'] == 'MS']
```

```python
grp_g1_ms
```

```python
grp_g2_ms = grp_g2.loc[grp_g2['school'] == 'MS']
```

```python
# Combining data on common columns
best_grades = pd.merge(grp_g1_ms, grp_g2_ms)
```

```python
best_grades
```

```python
# Combining dataframes across rows
ms_grades = pd.concat([grp_g1_ms, grp_g2_ms])
ms_grades
```

### Apply function

```python
# Get the average of the maximum grades
best_grades['grades_avg'] = best_grades[['G1_max', 'G2_max']].apply(np.average, axis=1)
best_grades
```

```python
# Change float type to integer
best_grades['grades_avg'] = best_grades['grades_avg'].astype(int)
best_grades
```

### For loop

```python
# Transform grades_avg into a binary format by iterating each row
result = []
for i in range(len(best_grades)):
    if best_grades['grades_avg'][i] >= 15:
        result.append('High')
    else:
        result.append('Low')
best_grades['grades_binary'] = result
```

```python
best_grades
```

```python
# OR
best_grades['grades_binary_'] = ['High' if grade >=15 else 'Low' for grade in best_grades['grades_avg']]
best_grades
```

### Functions

```python
# Define a function that averages high grades by gender, returns a list of tuples
def avg_high_grades(df):
    high_grades = df.loc[df['grades_binary'] == 'High']
    high_grades_grp = high_grades.groupby('sex')['grades_avg'].mean()
    # print(high_grades_grp)
    return list(zip(high_grades_grp.index, high_grades_grp))
```

```python
avg_high_grades(best_grades)
```

```python
# Define a function that takes two values, returns 1 value
def concatenate(value_1, value_2):
    return str(value_1)+ "--" + str(value_2)
```

```python
best_grades['concatenated'] = best_grades[['age','sex']].apply(
    lambda row: concatenate(row['age'], row['sex']) , axis=1)
```
